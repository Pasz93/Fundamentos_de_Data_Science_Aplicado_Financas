```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

# Carregar os dados
fonte = 'https://raw.githubusercontent.com/daniel-usp/LIT/master/500MM.csv'
df = pd.read_csv(fonte, sep=';', decimal=',')
df = df.drop('Empresa', axis=1)

# Padronizar os dados
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Rodar o K-means com três grupos
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Mostrar os resultados dos centroides
centroids = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=df.columns[:-1])
print("Centroides dos Grupos:")
print(centroids)

# Inserir os labels dos clusters no DataFrame original
df['Cluster'] = df['Cluster'].astype(str)

# Plotar o método do cotovelo para clusters de 1 a 5
inertia = []
for i in range(1, 6):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plotar o método do cotovelo
plt.plot(range(1, 6), inertia, marker='o')
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('Inércia')
plt.show()

# Visualizar o DataFrame com os labels dos clusters
display(df)
```
